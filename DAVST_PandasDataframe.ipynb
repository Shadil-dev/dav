To create a Dataframe

```python
import pandas as pd

data = {
    "calories": [420, 380, 390],
    "duration": [50,40,45]
        }

df=pd.DataFrame(data)

print(df)
```

To locate the row in a dataframe

```python
print(df.loc[[0,1,2]])
```

Named Index

```python
df=pd.DataFrame(data, index=["day1", "day2", "day3"])
print(df)
```

To locate named Index

```python
print(df.loc["day2"])
```
